```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
data=pd.read_csv('breast_cancer.csv')
data.head(2)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
classifier=LogisticRegression(penalty='l2',C=1.0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
confusion_matrix(y_pred,y_test)
```

```python
import math as mt
acc=accuracy_score(y_pred,y_test)
print("The model has achieved accuracy of:- {}%".format(mt.ceil(acc*100)))
```

```python
#Computing with cross_val score
from sklearn.model_selection import cross_val_score,cross_val_predict
accuracies=cross_val_score(estimator=classifier, X=x_train, y=y_train,cv=10)
print("Average accuracy of the model after cross validation is:- {:.2f} %".format(accuracies.mean()))
print("Standard deviation of the model is:- {:.2f} %".format(accuracies.std()*100))
```
